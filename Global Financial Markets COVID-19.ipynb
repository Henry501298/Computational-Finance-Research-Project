```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')

tickers = ['SPY', 'EWU', 'EWG', 'MCHI', 'INDA']
ind_data = pd.DataFrame()
for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source='yahoo', start='2020-2-20', end='2020-6-23')['Adj Close']
```

SPY = SPDR S&P 500 ETF

EWU = iShares MSCI United Kingdom ETF 

EWG = iShares MSCI Germany ETF

MCHI = iShares MSCI China ETF

INDA = iShares MSCI India ETF

Global Financial Markets ETFs performance since COVID-19:

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize=(20, 10));
plt.show()
```

### Chart Analysis: Since the impact of COVID-19 on global financial markets ETFs, we could notice out of these ETFs, MCHI (China ETF) is the top performing ETF.

Overall daily returns of the Global Financial Markets ETFs since COVID-19:

```python
ind_returns = (ind_data / ind_data.shift(1)) - 1
ind_returns.tail()
```

Overall annual return of the Global Financial Markets since COVID-19:

```python
annual_ind_returns = ind_returns.mean() * 252
annual_ind_returns
```

### Return Analysis Since COVID-19: SPY -5.51%, EWU -42.55%, EWG 0.15%, MCHI 19.90% and INDA -21.17%

## Global Financial Markets ETFs Before COVID-19 Returns

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')

tickers = ['SPY', 'EWU', 'EWG', 'MCHI', 'INDA']
ind_data = pd.DataFrame()
for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source='yahoo', start='2019-1-20', end='2020-2-20')['Adj Close']
```

Global Financial Markets ETFs performance before COVID-19:

### Chart Anaysis: Before the impact of COVID-19 we could notice that SPY was the top performing ETF and MCHI was trailing behind

Overall daily returns of the Global Financial Markets ETFs before COVID-19:

Overall annual return of the Global Financial Markets before COVID-19:

```python
annual_ind_returns = ind_returns.mean() * 252
annual_ind_returns
```
