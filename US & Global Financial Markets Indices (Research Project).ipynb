## United States Indicies 

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')

tickers = ['^GSPC', '^DJI', '^IXIC']
ind_data = pd.DataFrame()
for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source='yahoo', start='2008-1-1')['Adj Close']
```

## Indices Performances From 2008 - 2020 

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize=(20, 10));
plt.show()
```

### From 2008 - 2020 the US Markets have seen an enormous increase in value and size in the last 12 years. With 2008 - 2009 being the worse performing years because of the "great recession" caused by the housing market collaspe do to banks lending out money to people who really just could not afford it at that time. Despite what happen in 2008 - 2009, you can notice that US Markets started to rally into a "bull market" in late 2009 and has been on an upward trend since 2020, causing it to be one of the longet bull markets in history for the U.S. Stock market. Now do to COVID-19, we have experienced a much larger drop in the market compared to the "great recession". It is still too early to tell if we are going to enter a recession, however it is highly possible we may enter a recession with the Federal Reserve printing money and setting interest rates to 0% causing inflation to increase. 

### throughout the past 12 years, we can notice that the Nasdaq index (^IXIC) has been outperforming the S&P 500 and Dow Jones Index funds. The S&P 500 is considered the benchmark for the U.S. Market, and with Nasdaq outperforming the S&P 500, brings optimism for the future.  From this chart we can take notice that the S&P 500 index and the Dow Jones Index have a very close correlation with performances, however the S&P 500 still remains in the lead.

Indices Returns:

```python
ind_returns = (ind_data / ind_data.shift(1)) - 1
ind_returns.tail()
```

```python
annual_ind_returns = ind_returns.mean() * 252
annual_ind_returns
```

### At the time of me creating this assignment, the following annual returns are subjected to change. Which could lead to a higher or lower values. The S&P 500 index has currenly reached an annual return of 8%, Dow Jones of 7% and Nasdaq reaching annual return of 13%.

## Global Financial Markets Compared to the U.S. S&P 500 Index

```python
tickers = ['^GSPC', '^GDAXI', '000001.SS']
ind_data = pd.DataFrame()
for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source='yahoo', start='2008-1-1')['Adj Close']
```

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize=(20, 10));
plt.show()
```

### This Graphs compares the S&P 500 Index ( U.S. Benchmark, ^GSPC) and the German Index (Ger. Benchmark, ^GDAXI) and the Shangahi Index (Ch. Benchmark, 000001.SS). From this graph we can notice that the S&P 500 index has been surpasing both the German and Shangahi indices, with the Shangahi index fall way behind both the German and S&P 500 index. We can also take note that the S&P 500 index and the German Dax Index have a similar correlation movement, however it still falls behind the S&P 500.
