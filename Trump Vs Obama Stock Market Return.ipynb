## Obama Administraion

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')

tickers = ['^GSPC', '^DJI', '^IXIC']
ind_data = pd.DataFrame()
for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source='yahoo', start='2009-1-20', end='2017-1-20')['Adj Close']
```

^GSPC = S&P 500 Index

^DJI = Dow Jones Index 

^IXIC = Nasdaq Index

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize=(20, 10));
plt.show()
```

Overall daily returns of the Obama Administration from 2009 to 2017:

```python
ind_returns = (ind_data / ind_data.shift(1)) - 1
ind_returns.tail()
```

Overall annual return of the Obama Administration from 2009 to 2017

```python
annual_ind_returns = ind_returns.mean() * 252
annual_ind_returns
```

### Overall Obama Administration Analysis: The S&P 500 reached 14.44%, Dow Jones 12.68% and Nasdaq 18.66%
